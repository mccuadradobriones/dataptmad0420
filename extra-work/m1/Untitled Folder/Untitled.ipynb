```python
import pandas as pd
```

```python
df=pd.read_csv(r'/home/carmencuadrado/Ironhack/dataptmad0420/extra-work/m1/Untitled Folder/IMDB-Movie-Data.csv')
```

```python
##Challenge 1. Warm up

##We want to create bins of movies according to the number of votes they've received. For that matter, we will create a new column named 'bin' which will tag every movie as follow:

    #From 0 to 999 ==> 1
    #From 1000 to 9999 ==> 2
    #From 10000 to 99999 ==> 3
    #From 100000 to 999999 ==> 4
    #More than 1000000 ==> 5
```

```python
df['Bins']=df['Votes'].apply(lambda x:1 if x<=999 else(2 if x<=9999 else (3 if x<=99999 else (4 if x<=999999 else 5))))
```

```python
df['Bins'].unique()
```

```python
#Challenge 2. Using axis concept
#We want to know how much is the revenue per minute for every movie.
```

```python
df['Revenue per minute']=df.apply(lambda x: x['Revenue (Millions)']/x['Runtime (Minutes)'], axis=1)
```

```python
#Challenge 3. Using the lambda
#We want to create a new ranking where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.
```

```python
df['New_Rank_by_Genre'] = df.apply(lambda x : x['Rank']-1 if 'Comedy' in x['Genre'] else (x['Rank']+1 if 'Thriller' in x['Genre'] else x['Rank']), axis =1)
```

```python
#Challenge 4. Now the real stuff
#We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers.
```

```python
df['ascii/value']=df.apply(lambda x: sum([(ord(i)) for i in x['Title']])/x['Votes'], axis=1)
```

```python
df[df['ascii/value']>=1]['ascii/value'].unique()
```

```python
print('Diving ascii summed values for Titles by Votes yields float numbers. Selecting those who are =>1 I do not see any that could be potentially converted to int, though I conclude that there is not prime numer, as prime numbers are int.') 
```

```python
#Challenge 5. And finally some fantasy
#Feel free to propose your own ranking based in aggregations of at least 3 columns of the dataset.
```

```python
new_ranking=df[['Title','Rating', 'Votes', 'Revenue (Millions)']].sort_values(['Rating', 'Votes', 'Revenue (Millions)'],ascending=False).reset_index()
new_ranking.index=pd.RangeIndex(len(new_ranking.index))
new_ranking=new_ranking.assign(New_Ranking_by_Votes_Rating_Revenue=new_ranking.index.get_level_values(None))
```

```python
new_ranking.drop(columns='index')
df=pd.merge(df,new_ranking, left_on='Title', right_on='Title', how='left')
df.drop(columns=['index','Rating_y','Votes_y','Revenue (Millions)_y'], inplace=True)
```
